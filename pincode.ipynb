```python
df_full['narration'] = df_full['narration'].str.lower()
df_full['narration'] = df_full['narration'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
display(df_full.head())
```

```python
df_full['narration'] = df_full['narration'].str.lower()
df_full['narration'] = df_full['narration'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
display(df_full.head())
```

```python
import sys
!{sys.executable} -m pip uninstall -y fuzzywuzzy > /dev/null
!{sys.executable} -m pip install rapidfuzz > /dev/null
from rapidfuzz import fuzz, process
print("Rapidfuzz is installed and imported.")
```

```python
narration_matching_results = []

# Get unique IFSC codes (already lowercased in df_i['IFSC'])
unique_ifsc_codes = df_i['IFSC'].unique()

# Get the narration column as a list (already preprocessed in df_full['narration'])
narration_list = df_full['narration'].tolist()

# Iterate through each narration
for idx, narration in enumerate(narration_list):
    # Find the best matching IFSC code in the unique_ifsc_codes list
    best_match = process.extractOne(narration, unique_ifsc_codes, scorer=fuzz.token_set_ratio)

    if best_match:
        matched_ifsc = best_match[0]
        similarity_score = best_match[1]
        narration_matching_results.append({
            'Original_Narration_Index': idx,
            'Original_Narration': narration,
            'Matched_IFSC': matched_ifsc,
            'Similarity_Score': similarity_score
        })

# Convert the results list to a pandas DataFrame
df_narration_matches = pd.DataFrame(narration_matching_results)

# Display the head of the new DataFrame
display(df_narration_matches.head())
print(f"Total narration matches found: {len(df_narration_matches)}")
```

```python
matching_results = []

# Get unique IFSC codes to avoid redundant processing
unique_ifsc_codes = df_i['IFSC'].unique()

# Get the narration column as a list for process.extractOne efficiency
narration_list = df_full['narration'].tolist()

# Iterate through each unique IFSC code
for ifsc_code in unique_ifsc_codes:
    # Find the best match in the narration column
    # scorer=fuzz.ratio is a common choice for general string similarity
    best_match = process.extractOne(ifsc_code, narration_list, scorer=fuzz.ratio)

    if best_match:
        matched_narration = best_match[0]
        similarity_score = best_match[1]
        matching_results.append({
            'Original_IFSC': ifsc_code,
            'Matched_Narration': matched_narration,
            'Similarity_Score': similarity_score
        })

# Convert the results list to a pandas DataFrame
df_matches = pd.DataFrame(matching_results)

# Display the head of the new DataFrame
display(df_matches.head())
```

# RTGS code

```python
df_rtgs = pd.read_excel('/content/drive/MyDrive/live_project/rtgs.xlsx')
```

```python
df_rtgs.head()
```

```python
rtgs = df_rtgs['Code'].dropna().to_list()
print(len(rtgs))
print(rtgs[:5])
```

```python
import re

# Create a regex pattern to match any of the RTGS codes
# Ensure all codes are converted to strings before calling .lower()
rtgs_pattern = '|'.join(re.escape(str(code).lower()) for code in rtgs)

# Count how many narrations contain any of the RTGS codes
rtgs_in_narration_count = df_full['narration'].str.contains(rtgs_pattern, na=False, regex=True).sum()

print(f"Number of rows in df_full['narration'] containing an RTGS code: {rtgs_in_narration_count}")
```

```python
df_rtgs_matches = df_full[df_full['narration'].str.contains(rtgs_pattern, na=False, regex=True)]
display(df_rtgs_matches.head())
```

```python
import sys
!{sys.executable} -m pip install spacy > /dev/null
print("spaCy installed.")
```

```python
from spacy import displacy
import re

# Prepare data for displacy
displacy_data = []

for index, row in df_rtgs_matches.head(10).iterrows(): # Limiting to first 10 for demonstration
    narration = row['narration']
    ents = []

    # Find all occurrences of RTGS codes in the narration
    for match in re.finditer(rtgs_pattern, narration, re.IGNORECASE):
        start, end = match.span()
        ents.append({"start": start, "end": end, "label": "RTGS_CODE"})

    displacy_data.append({"text": narration, "ents": ents, "title": None})

# Define options for displacy rendering
options = {"colors": {"RTGS_CODE": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}}

# Render the data
displacy.render(displacy_data, style="ent", manual=True, jupyter=True, options=options)
```

## Identify and Display Matched Locations

### Subtask:
Iterate through each narration in the `df_location_matches` DataFrame, use the `location_pattern` to find all matched location substrings, identify the exact location name from the `locationlist` that corresponds to it (case-insensitively), and print the original narration, the matched substring, and the identified location name from the `locationlist`.


**Reasoning**:
I will implement the instructions to iterate through `df_location_matches`, find location substrings using `re.finditer` and `location_pattern`, identify the exact location name from `locationlist` that corresponds to each match, and store these in a new DataFrame for display.



```python
import re
import pandas as pd

# 1. Initialize an empty list to store results
location_tagging_results = []

# 2. Iterate through each row of the df_location_matches DataFrame
for index, row in df_location_matches.iterrows():
    original_narration = row['narration']

    # 3. Use re.finditer() with location_pattern to find all occurrences of locations
    for match in re.finditer(location_pattern, original_narration, re.IGNORECASE):
        # 4a. Extract the matched_substring
        matched_substring = match.group(0)

        # 4b. Iterate through the locationlist to find the exact matched_location
        matched_location = None
        for loc in locationlist:
            if str(loc).lower() == matched_substring.lower():
                matched_location = loc # Store the original cased location from list
                break

        # 4c. Append a dictionary to location_tagging_results
        if matched_location:
            location_tagging_results.append({
                'Original_Narration': original_narration,
                'Matched_Substring': matched_substring,
                'Identified_Location': matched_location
            })

# 5. Convert the location_tagging_results list into a pandas DataFrame
df_location_tagging_results = pd.DataFrame(location_tagging_results)

# 6. Display the head of the newly created DataFrame
display(df_location_tagging_results.head())
print(f"Total unique narrations with identified locations: {df_location_tagging_results['Original_Narration'].nunique()}")
print(f"Total location matches found: {len(df_location_tagging_results)}")
```
