# IFSC codes

We get bank names , IFSC code and rtgs code(the first 4 characters of an ifsc code).

```python
df_i = pd.read_excel('/content/drive/MyDrive/live_project/ifsc.xlsx')
display(df_i.head())
print(df_i.info())
```

```python
df_i['IFSC'] = df_i['IFSC'].str.lower()
display(df_i.head())
```

```python
df_full['narration'] = df_full['narration'].str.lower()
df_full['narration'] = df_full['narration'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
display(df_full.head())
```

```python
df_full['narration'] = df_full['narration'].str.lower()
df_full['narration'] = df_full['narration'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
display(df_full.head())
```

```python
import sys
!{sys.executable} -m pip uninstall -y fuzzywuzzy > /dev/null
!{sys.executable} -m pip install rapidfuzz > /dev/null
from rapidfuzz import fuzz, process
print("Rapidfuzz is installed and imported.")
```

```python
narration_matching_results = []

# Get unique IFSC codes (already lowercased in df_i['IFSC'])
unique_ifsc_codes = df_i['IFSC'].unique()

# Get the narration column as a list (already preprocessed in df_full['narration'])
narration_list = df_full['narration'].tolist()

# Iterate through each narration
for idx, narration in enumerate(narration_list):
    # Find the best matching IFSC code in the unique_ifsc_codes list
    best_match = process.extractOne(narration, unique_ifsc_codes, scorer=fuzz.token_set_ratio)

    if best_match:
        matched_ifsc = best_match[0]
        similarity_score = best_match[1]
        narration_matching_results.append({
            'Original_Narration_Index': idx,
            'Original_Narration': narration,
            'Matched_IFSC': matched_ifsc,
            'Similarity_Score': similarity_score
        })

# Convert the results list to a pandas DataFrame
df_narration_matches = pd.DataFrame(narration_matching_results)

# Display the head of the new DataFrame
display(df_narration_matches.head())
print(f"Total narration matches found: {len(df_narration_matches)}")
```

```python
matching_results = []

# Get unique IFSC codes to avoid redundant processing
unique_ifsc_codes = df_i['IFSC'].unique()

# Get the narration column as a list for process.extractOne efficiency
narration_list = df_full['narration'].tolist()

# Iterate through each unique IFSC code
for ifsc_code in unique_ifsc_codes:
    # Find the best match in the narration column
    # scorer=fuzz.ratio is a common choice for general string similarity
    best_match = process.extractOne(ifsc_code, narration_list, scorer=fuzz.ratio)

    if best_match:
        matched_narration = best_match[0]
        similarity_score = best_match[1]
        matching_results.append({
            'Original_IFSC': ifsc_code,
            'Matched_Narration': matched_narration,
            'Similarity_Score': similarity_score
        })

# Convert the results list to a pandas DataFrame
df_matches = pd.DataFrame(matching_results)

# Display the head of the new DataFrame
display(df_matches.head())
```

# RTGS code

```python
df_rtgs = pd.read_excel('/content/drive/MyDrive/live_project/rtgs.xlsx')
```

```python
df_rtgs.head()
```

```python
rtgs = df_rtgs['Code'].to_list()
print(len(rtgs))
print(rtgs[:5])
```

```python
import re

# Create a regex pattern to match any of the RTGS codes
# Ensure all codes are converted to strings before calling .lower()
rtgs_pattern = '|'.join(re.escape(str(code).lower()) for code in rtgs)

# Count how many narrations contain any of the RTGS codes
rtgs_in_narration_count = df_full['narration'].str.contains(rtgs_pattern, na=False, regex=True).sum()

print(f"Number of rows in df_full['narration'] containing an RTGS code: {rtgs_in_narration_count}")
```

```python
df_rtgs_matches = df_full[df_full['narration'].str.contains(rtgs_pattern, na=False, regex=True)]
display(df_rtgs_matches.head())
```

```python
def tag_matched_rtgs(narration, pattern):
    # Ensure narration is a string to avoid errors with non-string types
    if not isinstance(narration, str):
        return narration

    # Use re.sub to replace all occurrences of the pattern with a tagged version
    # The 'flags=re.IGNORECASE' is important since the pattern was made lowercase.
    tagged_narration = re.sub(f'({pattern})', r'<MATCHED_RTGS>\1</MATCHED_RTGS>', narration, flags=re.IGNORECASE)
    return tagged_narration

# Explicitly create a copy to avoid SettingWithCopyWarning
df_rtgs_matches = df_full[df_full['narration'].str.contains(rtgs_pattern, na=False, regex=True)].copy()
df_rtgs_matches['tagged_narration'] = df_rtgs_matches['narration'].apply(lambda x: tag_matched_rtgs(x, rtgs_pattern))
display(df_rtgs_matches[['narration', 'tagged_narration']].head())
```
